# General principles of energy conservation

## Potential and kinetic energy in 1D

Consider one dimensional motion under a force that is a function of $x$ **only** (constant forces okay too).
No position, no time.
We saw this case in lecture 1,
$$F(x)=ma=m\ode{v}{t} = m\ode{v}{x}\ode{x}{t}=m v \ode{v}{x},$$
We had essentially removed the variable time from the derivative and were able to use separation of variables,
$$mv\di v = F(x)\di x \thrfor m\int_{v_0}^v v'\di v' = \int_{x_0}^x F(x')\di x'$$
\begin{equation}
\thrfor\frac12 m\left[v'^2\right]_{v_0}^v = \int_{x_0}^x F(x')\di x' \thrfor \frac12 m \left(v^2 -  v_0^2\right)= \int_{x_0}^x F(x')\di x'.
\label{eq:WorkEnergy}
\end{equation}
You might have recognized this as a standard Work/Kinetic energy equation, and we will dig deeper now.

If we now define a function $U(x)$ such that 
$$F(x)= -\ode{U}{x},$$ 
we can find $U(x)$ for any $F(x)$ using
\begin{equation}
    U(x) = -\int_{x_0}^x F(x')\di x' + U_0,\with U(x_0)=U_0.
    \label{eq:Vdef}
\end{equation}
We can then isolate for the integral of $F$ in this equation and plug it into eqn. \eqref{eq:WorkEnergy},
$$\frac12 m \left(v^2 -  v_0^2\right) = U_0 - U(x) \thrfor \frac12 m v^2 + U(x) =  \frac12 m v_0^2 + U_0.$$
Since the RHS is a constant, and the LHS is true for an arbitrary $x$, it must apply for *any* $x$.
This means that the LHS of the equation is always equal to a constant, it is **conserved**.

We define the **mechanical energy** as
$$E=\frac12 mv^2 + U(x) = K(\dot x)+ U(x),$$
where $U(x)$ is the **potential energy (PE)** and $K=\frac12 mv^2$ is the **kinetic energy (KE)**.

## Work

Let's go back to eqn. \eqref{eq:WorkEnergy} and see how the concept of work fits into it.
Allow me to go back to $n$ dimensions instead of one for a sec: the infinitesimal work $\delta W$ done by a force $\vec F$ on an object or system during an intifinitesimal displacement $\di \vec r$ is $\delta W = \vec F\cdot\di\vec r$.
So, if the force $\vec F$ points in the same direction as the motion $\di \vec r$, then the work is positive (force increases the energy of the object or system), and vice-versa.

The *total* work done by a force during a motion going from $\vec r_0$ to $\vec r$ is $W = \int^{\vec r}_{\vec r_0} \vec F\cdot \di \vec r'$.
Now, if the force depended on e.g., time $t$, the total work would depend on time as well: say, the force is very weak during a certain time, and a particle goes from $\vec r_0$ to $\vec r$ during that time, the work due to that force on the motion will be very small; it may also depend on the path taken: if, at constant speed, the trajectory describes loops on the way from $\vec r_0$ to $\vec r$, and the strength of the force evolves in time, the work might not be the same as if the motion went in a straight line, which is more direct and more likely to sample the value, the force had, during this short time.

When $\vec F$ can be written as $-\grad U$, however, where this object is a vector whose Cartesian decomposition reads
$$\grad{U} = \pde Ux\unit x + \pde Uy \unit y + \pde Uz \unit z,$$
**the work done by $\bs{\vec F}$ only depends on the initial and final positions:**
$$W = \int_{\vec r_1}^{\vec r_2} -\grad U\cdot \di\vec r = -U(\vec r_2) + U(\vec r_1).$$
This full 3D result is called the *gradient theorem*.
It is essentially a generalization of the fundamental theorem of calculus to multiple dimensions, and is the focus of PHY250 (Electricity and Magnetism) more than of PHY254.
Rest assured, our present course will only deal with 1D cases, such as the little derivation we did in the previous section.
If we ever deal with 2- or 3D cases, however, we will make sure to go gently.

So, when we can write $\vec F = -\grad U$, the work does not depend on the shape of the path, not on the time of day, not on how fast we get there, nor anything else.
And the change in KE only depends on the change in PE, which itself only depends on the initial and final positions.

A word on the signs: if the KE difference between the initial and final moments of a motion is $\Delta K$, and the PE difference during the same motion is $\Delta U$, and the energy is conserved ($\Delta K + \Delta U = 0$), then the quantities are related by $\Delta K = -\Delta U = W$: positive work by a force tends to accelerate a motion, negative work by a conservative force tends to store PE in the system.

Finally, about the sign of $W$: many references talk about work done done by *an operator* (e.g., you) *against* a force.
If you took or will take PHY250, you will use the textbook by Griffiths, and he uses this convention. (See his §2.4.1, p. 91.)
It is the opposite to the convention I used in this section, where I calculated the work done *by* the force.
Using one or the other flips the sign of the work ($W = \pm [U(\vec r_2) - U(\vec r_1)]$), but the final result should be the same: lifting an object against gravity stores potential energy in the object, and so does stretching a spring.

## Summary: what makes a force conservative?

In our little explanation, we made two assumptions about the force which led to the principle of mechanical energy conservation, namely

1. the force $\vec F$ only depends on position $\vec r$,
2. the force $\vec F$ can be written as minus the gradient of a potential energy, $\vec F = -\grad{U}$.
    The gradient theorem tells us that it is equivalent to writing that the work done by $\vec F$ during the motion only depends on the initial and final positions, and not on the path or the time of day.

If you have these two conditions met, then you can come up with an energy conservation equation like that of Eq. \eqref{eq:WorkEnergy}, and **the force is conservative**.
Conversely, **we can only define a potential energy for forces that are conservative**.

Those two assumptions are independent indeed.
In 1D, imagine a scalar field $U$ that depended on position, but also oscillated in time, $U(x, t) = \overline U(x)\cos(\omega t)$.
Then, you could define the force as $F = -\partial U/\partial x = -(\odel{\overline U}x)\cos(\omega t)$, but it would have a time-dependence $\cos(\omega t)$.
So, condition 2 does not necessarily imply condition 1.

Could condition 1 imply condition 2? In 1D, it is hard to imagine a force that would depend on $x$ only and not be the derivative of a spatial function, simply because you can always define this function with equation \eqref{eq:Vdef}.
This is true indeed that in 1D, condition 1 always implies condition 2. 
But in 2D or 3D, this isn't always the case, and you can have so-called *rotational forces*, which only depend on position, but derive from the *curl* of a (vector) potential rather than the *gradient* of one... But in PHY254, we will not worry about such cases.

We will mostly work in 1D, so remember this: **In 1D, a force that only depends on position is conservative**.

## A method to solve problems with energy conservation

We can find an explicit solution for any energy conserving system: since
$$\frac12 mv^2=E-U(x)$$
(remember that $E$ is a constant) then
$$v(x)=\pm\sqrt{\frac 2m\left[E-U(x)\right]}=\dot x.$$

This is a first order ODE, unlike Newton's law, which is second order! (Of course, this is because we already integrated once to get the energy equation).

Using separation of variables, we can find
$$t-t_0=\pm\int_{x_0}^x\frac{\di x'}{v(x')}=\pm\int_{x_0}^x\frac{\di x'}{\sqrt{\frac 2m\left[E-U(x')\right]}}.$$ From this we can obtain $t=t(x)$.  Also, if the expression is not too gruesome, we can invert it to find $x(t)$, as we are about to see with oscillators.

# Energy conservation in oscillators

## Simple harmonic motion (SHM)

If $F(x)=-kx$, where $x$ is the departure of the spring from equilibrium, then using equation \eqref{eq:Vdef} we find
$$U(x)=\frac12 kx^2$$
and the mechanical energy is
$$E=\frac12 m\dot x^2+\frac12 kx^2.$$
Note that we had to define $x=0$ the position at which $U=0$.
This was convenient, but it was not imperative to do so: potential energies are always measured with respect to an arbitrarily-defined level.
When we write that the gravitational potential energy of a mass $m$ at altitude $h$ is $mgh$, we implicitly mean "with respect to the ground".
And when we define that the gravitational potential energy of a two-mass system to be $-G m_1 m_2/r$, we implicitly mean "with respect to the state where they are separated by infinite distance".
The arbitrary level is usually the one for which you can't turn any PE into KE, which is the case for the three situations mentioned in this paragraph: a mass+spring at rest, a mass that fell to the ground, and two infinitiely far masses won't suddenly start moving due to an excess of PE.

Back to our spring: if $-kx$ is the only force acting, then we have SHM with angular frequency $\omega_0=\sqrt{k/m}$ (old news by now).
Notice that this is actually the ratio of the coefficients of the KE and PE terms. In fact, we can write $$E=\frac12 m\left(\dot x^2+\omega_0^2 x^2\right).$$

In general, if energy is conserved and we can write its equation in the form
\begin{equation}
    E = a\dot q^2 +  b q^2,
    \label{eq:GeneralPot}
\end{equation}
where $q$ is some variable related to position and $E$ is constant, we know that the solution will be SHM, and that frequency will satisfy $\omega^2=b/a$.
Indeed,
$$\dot{E} = 2a \dot q \ddot q + 2b q \dot q = 2\left(\ddot q + \frac{b}{a}q \right)a\dot q =0,$$
and because $a\neq 0$, $\dot q \neq 0$ in general, then the expression within the parentheses is zero, which is the SHO equation with natural angular frequency $\sqrt{b/a}$.

We could have retrieved conservation of energy from the SHM solution we derived in lecture 3.
Let's consider a SHM of amplitude $A$ and define a time origin for which the spring extent is maximum, namely,
$$x(t) = A\cos(\omega_0 t) \andeq v(t) = \dot x(t) = -\omega_0 A\sin(\omega_0 t).$$
The mechanical energy is then, using $k = m\omega_0^2$,
\begin{align*}
E & = \underbrace{\frac12 m\omega_0^2 A^2\cos^2(\omega_0 t)}_{U = kx^2/2} + \underbrace{\frac12 m A^2\omega_0^2 \sin^2(\omega_0 t)}_{K = mv^2/2} \\
  & = \frac12 m\omega_0^2 A^2 \left(\cos^2(\omega_0t) + \sin^2(\omega_0 t)\right) = \frac12 m\omega_0^2 A^2,
\end{align*}
where I used $\cos^2 + \sin^2 = 1$.
So, we do have $E = m\omega_0^2 A^2/2$ constant.

Plotting $K$ and $U$ individually highlights "energy juggling", the fact that during SHM, KE and PE constantly trade in a zero-sum game.

```python
# Let us plot this.
def plot_SHO_E(k, m, A):
    """ Function that computes and plots the energy
    INPUT:
    k: spring stiffness [N/m, kg/(s**2)]
    m: mass [kg]
    x0: initial position [m]
    initial velocity is treated as constant
    OUTPUT: plot of the energies"""
    
    omega0 = (k/m)**.5  # [rad/s] angular frequency
    T0 = 2*np.pi/omega0  # [s] period
    
    t = np.linspace(0., 0.6, 128)  # time array, from 0 to 0.6 s, 128 points
    x = A*np.cos(omega0*t)  # position [m]
    v = -A*omega0*np.sin(omega0*t)  # [m/s] velocity
    
    K = 0.5*m*v**2  # [J or kg*(m/s)**2] kinetic energy
    U = 0.5*k*x**2  # [J or kg*(m/s)**2] potential energy
    E = 0.5*k*A**2  # [J or kg*(m/s)**2] mechanical energy
    
    ftsz = 13  # font size

    plt.figure()
    ax1 = plt.gca()
    ax1.plot(t, K, 'b', label='$K$')  # plotting the KE K
    ax1.plot(t, U, 'r--', label='$U$')  # plotting the PE U
    ax1.legend()
    ax1.set_xlabel('time [s]', fontsize=ftsz) 
    ax1.set_ylabel(r'Energy [J, or kg.m$^2$.s$^{-2}$]', color='b', fontsize=ftsz)

    # annotation to highlight the total energy
    ax1.axhline(E, color='k', linestyle='-.')  # total mechanical energy
    ax1.text(T0*1.05, E*0.97, '$E = m\omega_0^2A^2/2$', color='k',
             verticalalignment='center', horizontalalignment='left',
             backgroundcolor='w', fontsize=ftsz)

    ax1.axhline(0., color='k')  # draw the zero-axis as horizontal line
    ax1.set_xlim([t.min(), t.max()])

    # annotation to highlight the period
    ax1.axvline(T0, color='k', linestyle='-.')  # the t=T mark
    ax1.annotate(text='', xy=(0., 0.2*E), xytext=(T0, 0.2*E),
                 arrowprops=dict(arrowstyle='<|-|>'))  # the double arrow
    ax1.text(T0/2, 0.2*E, r'$T_0 = 2\pi/\omega_0 \approx {0:.2f}$ s'.format(T0),
             verticalalignment='center', horizontalalignment='center',
             backgroundcolor='w', fontsize=ftsz)
    return
```

```python
plot_SHO_E(k=180., m=0.8, A=4e-2)  # plot for pdf export
```

Couple of things to notice from the plot and analytical solutions:

* Energy goes as the square of the amplitude. Sort of obvious when you look at the definitions, but it is interesting to keep in mind that doubling the amplitude of a motion quadruples its energy.
* The frequency in the energy juggling is twice the frequency of the motion.
    Also makes sense, but worth thinking about it for a minute.
    At maximum extension, $K=0$ and $U = kx_0^2/2$ is maximum.
    When the mass passes by the rest position $x=0$ at $t=T/4$, the velocity and therefore $K$ are maximum, but the spring is neither stretched or compressed and $U=0$. 
    Half a period later, the spring is at its maximum compression (if initially stretched), $U = kx_0^2/2$ again, and the velocity is zero ($K=0$ again).
    So, during one oscillation cycle, the KE is maximum twice, and so is the PE, but not at the same time.
    The periodicity of $K$ and $U$ is twice that of $x$ and $v$, which one can tell mathematically because $\sin^2(\omega t) = [1 - \cos(2\omega t)]/2$ and $\cos^2(\omega t) = [1 + \cos(2\omega t)]/2$.

## Lightly-damped harmonic motion

We now have a non-conservative force in the mix, $F_d = -bv$.
It is not conservative because it depends on $v$ and not $x$, which in 1D means that the second condition is also automatically not met.
Let's see how to deal with it.

Whether the mass moves or is held in place, the friction does not represent an energy reservoir like KE or PE do: it merely removes mechanical energy from the system.
This is a physical expression of the fact that there is no point in defining a PE for a non-conservative force.
Therefore, the energy of the system has the same expression,
$$ E = K + U = \frac12 m \left(v^2 + \omega_0 x^2\right).$$

What the existence of friction/damping change is the *time evolution* of the mechanical energy of the \{mass+spring\} system: it decays because, well, it is not conserved anymore.

Let me repeat that I am not violating the conservation of total energy of a closed system.
This is because damping allows to exchange energy with other types of energy reservoirs.
For example, friction can create heat. 
The energy is still present in the system \{mass+spring+water\} if the damping medium is water, but it is in the form of *kinetic energy in the water* and *internal kinetic energy* (the stuff that makes temperature go up), which is not considered part of what defines mechanical energy.
But in any case, the *total energy of the closed system*, which may include the entire universe if need be, stays constant.
On the other hand, the system we are considering in this lecture \{mass+spring\} is *not* closed.

In the SHO, $E$ was constant because without any damping, the SHO is effectively a closed system, a system that cannot exchange energy with other reservoirs of energy, be they internal or external.

The damping does not represent an energy reservoir, but is does represent a leak, or a drain, of energy:

$$ \ode{E}t = \ode{}{t}\left(\frac12 m v^2 +\frac12k x^2\right) = mv\dot v + kx\dot x = (ma + kx)v, $$
and because $ma = -kx - bv$, 
$$ \boxed{\ode{E}t =  -bv^2}. $$
The mechanical energy loss is directly attributable to the damping.
Notice how the damping rate depends on the *square* of the linear quantities, here, $v^2$.
This is a reminder of the quadratic dependence of energy quantities versus linear quantities ("double the amplitude, quadruple the energy").

In a lightly-damped oscillator for which $\phi=0$, the solution is (see lecture 4 for detail)
$$x(t) = A_0\eexp{-\gamma t}\cos(\omega_d t)\andeq v(t) = -A_0 \eexp{-\gamma t} \left[\omega_d\sin(\omega_d t) + \gamma \cos(\omega_d t)\right],$$
with $A_0$ the initial amplitude and $\gamma$ the damping coefficient.

The potential energy is
$$ U = \frac12 m\omega_0^2 x^2 =  \frac12 m\omega_0^2 A_0^2 \cos^2(\omega_d t) \eexp{-2\gamma t}.$$

The kinetic energy is
\begin{align*}
    K & = \frac12 m A_0^2 \eexp{- 2\gamma t}\left[\omega_d\sin(\omega_d t) + \gamma\cos(\omega_d t)\right]^2 \\
      & = \frac12 m A_0^2 \omega_d^2 \eexp{- \gamma t}\left[\sin(\omega_d t) + \frac{\gamma}{\omega_d}\cos(\omega_d t)\right]^2 \\
      & \approx \frac12 m A_0^2 \omega_0^2 \eexp{- 2\gamma t}\sin^2(\omega_d t),
\end{align*}
where the last bit of approximation is valid when $\gamma \ll \omega_d$, which is equivalent to $\gamma\ll \omega_0$ or $Q \gg 1$.

The mechanical energy under the $Q \gg 1$ approximation is therefore
\begin{align*}
    E & = K + U \approx \frac12 A_0^2 \eexp{- 2\gamma t}m\omega_0^2\left[ \sin^2(\omega_d t) + \cos^2(\omega_d t)\right] \\
      & \approx \frac12 k A_0^2 \eexp{- 2\gamma t}.
\end{align*}

So, for a lightly damped oscillator, the mechanical energy decays exponentially.
The e-folding decay time scale $1/(2\gamma)$, or *half-a-lifetime* of the oscillation.

```python
 # let's plot, with the previous numerical values for the underdamped oscillator
def plot_DHO_E(k, m, Q, A0):
    """Same as plot_SHO_E but with damping this time"""
    ftsz = 13  # font size
    omega0 = (k/m)**.5  # [rad/s] angular frequency
    gamma = omega0/Q
    
    Q = omega0/gamma
    omegad = omega0*(1 - Q**-2)
    
    t = np.linspace(0., 2., 512)  # time array, from 0 to 0.6 s, 128 points
    x = A0*np.exp(-gamma*t)*np.cos(omegad*t)  # position [m]
    v = -A0*np.exp(-gamma*t)*(omegad*np.sin(omegad*t) + gamma*np.cos(omegad*t))
    v_approx = -A0*omega0*np.exp(-gamma*t)*np.sin(omegad*t)  # valid when Q>>1
    
    K = 0.5*m*v**2  # [J or kg*(m/s)**2] kinetic energy
    U = 0.5*k*x**2  # [J or kg*(m/s)**2] potential energy
    K_approx = 0.5*m*v_approx**2

    E = K + U
    E_approx = K_approx + U

    fig = plt.figure(dpi=100)
    ax1 = fig.gca()
    ax1.plot(t, K, 'b-.', label='$K(t)$')  # unapproximated kinetic energy
    ax1.plot(t, K_approx, 'b', label='$K(t)$, approx.')  # approximated kinetic energy
    ax1.plot(t, U, 'g--', label='$U(t)$')  # potential energy
    ax1.plot(t, E, 'r-.', label='$E(t)$')  # unapproximated mechanical energy
    ax1.plot(t, E_approx, 'r', label='$E(t)$, approx.')  # approximated mechanical energy
    ax1.set_xlim([0., 3.])
    ax1.set_xlabel('time [s]', fontsize=ftsz) 
    ax1.set_ylabel('Energies [J]', fontsize=ftsz)

    ax1.grid()
    ax1.axhline(0., color='k')  # draw the zero-axis as horizontal line

    plt.legend()
    plt.tight_layout()
    ax1.set_xlim(0., 2.)

    return
```

```python
plot_DHO_E(k=180., m=0.8, Q=10, A0=.2)
```

As you can see above, even for a moderately high $Q$, the approximation work OK.
This is the same sort of physics that happens: juggling between PE and KE, with the addition that total mechanical energy decays.

## Small angle pendulum

Based on what we learned from the spring, we should be able to find the frequency of the pendulum by finding the potential energy, writing out the mechanical energy equation and taking the ratio of the coefficients in front of the position and velocity terms.

For the kinetic energy, recall from lecture 3 that the velocity is along the tangent of the circle at any moment, and that its magnitude is $v_\theta = \ell \dot\theta$, and the kinetic energy is
$$K = \frac12 m v_\theta^2=\frac{1}{2}m(\ell\dot\theta)^2.$$

Setting the potential energy to zero when the pendulum is at $\theta=0$, and naming $h = \ell - \ell\cos\theta$ the height-above-rest-position (see sketch above), the gravitational potential energy is
$$U= mgh = mg\ell(1-\cos\theta).$$

For the small angle approximation, if we use a Taylor expansion of $\cos\theta \approx 1-\theta^2/2$ and get
$$U(\theta) \approx \frac12 mg\ell\theta^2.$$

So the mechanical energy is
$$E=\frac12 m\ell^2\dot\theta^2+mg\ell(1-\cos\theta) \approx \frac12 m \ell^2\dot\theta^2+\frac12 mg\ell\theta^2 = \frac12m\ell\left(\ell\dot\theta^2 + g\theta^2\right).$$ 
Recall that we can use the ratio of the coefficents for PE and KE to determine the natural angular frequency, so
$$\omega^2_0 = \frac{g}{\ell},$$
consistent with our previous results for SHM.

# Motion in a general potential

## What is a general potential?

We saw that when you have the form $U(x) \propto x^2$ (i.e., the potential is quadratic in position),  we could determine the natural frequency of a SHO by comparing the coefficient multiplying $x^2$ with the one multiplying $\dot x^2$ in the KE.
This result turns out to be useful in a more general context.
Suppose you have some general potential that does not have the $x^2$ form (think large-deviations pendulum, or Lennard-Jones potential), and recall that from conservation of energy $K+U = E$,
$$v=\pm\sqrt{\frac 2m\left(E-U(x)\right)}.$$ 

```python
%%html
<img src="Eplot.png" alt="Drawing" style="width: 400px;"/>
```

See sketch above for an arbitrary energy plot.
We can draw a few conclusions:

* Consider a particle starting between $x_A$ and $x_B$ with energy $E<U_c$, a shown.
    When the particle reaches $x_A$ or $x_B$, 
    $$E=\frac12 mv^2 + U(x) = U(x_A) = U(x_B),$$
    so $v=0$ and the direction of motion reverses.
    These are *turning points*. 
* By writing $v = \odel{x}t$ and then rearranging for $\di t = \di x/v$, we can integrate the equation to get an expression for the period of oscillation between $x_A$ and $x_B$ and back to $x_A$ (hence the factor 2),
    $$\tau_{AB} = 2\int_{x_A}^{x_B}\frac{\di x'}{\sqrt{\frac 2m\left[E-U(x')\right]}}.$$
* If the particle starts between $x_C$ and $x_D$, with the same energy $E$ as it had when it was between $x_A$ and $x_B$, it also oscillates, but with a different period $\tau_{CD}\ne\tau_{AB}$ since $U(x)$ is different here.
* If the particle starts with $E>U_c$, the motion will be unbounded with a single turning point on the left. 
* If a particle is located at some $x_0$ in a potential "valley" or "well" region, you can determine what initial speed the particle must have at $x_0$ to escape the potential well.
    This is defined as the *escape velocity*.
    For example, let's say we started somewhere between $x_A$ and $x_B$ in the graph above.
    Then in order to escape, our total energy must be greater than the value of the potential energy at the local maximum of the potential hill, i.e. $E > U_c$.
    This is because if $E>U_c$ then you will have some kinetic energy when you get to the top of the hill and hence you won't turn around and stay trapped in the well.
    We can write this condition explicitly for our initial position and the escape velocity we need as
    $$E=\frac12 mv_e^2+U(x_0)>U_c.$$ 
    Solve for $v_e$ and you find
    $$\frac12 mv_e^2=U_c-U(x_0)\rightarrow v_e=\sqrt{\frac 2m\left[U_c-U(x_0)\right]}.$$
    *Note: If we started a particle with velocity $v_e$ going to the left, it would still eventually escape, but it would first reach a turning point on the left and then start moving to the right before finally escaping.*

## When can we approximate the general potential as a quadratic potential?

We have seen that we could do it in the particular cases of the pendulum (L03 and above), and of the Lennard-Jones potential (T03 and now HW2).
In fact, these are very general results.
Consider a particle starting at rest at a point $x_0$, where $U$ is a minimum.
The force is zero there since $F=-\odel{U}x = 0$.

Now give the particle a small initial velocity, or displace it a little bit from equilibrium: the mechanical energy is now a little larger than the minimum PE. 
Yes, we uttered the word "small", meaning we are about to pull a Taylor.
If we expand $U(x)$ in a Taylor series for a small distance $\varepsilon$ from $x_0$ ($\varepsilon \ll x_0$), we get
\begin{align*}
U(x) & = U(x_0+\varepsilon) \approx U(x_0) + \varepsilon\underbrace{U'(x_0)}_{{}=0} + \frac12\varepsilon^2 U''(x_0).
\end{align*}
The second term is zero because $U$ is at a minimum at $x_0$.
For the kinetic energy, $K = m\dot x^2/2 = m\dot\varepsilon^2/2$ and so we have
$$E=\frac12 m\dot \varepsilon^2 + U(x_0) + \frac12 U''(x_0)\varepsilon^2,$$
as an equation for the approximate energy, which is conserved.

This is almost in the generalized form for a quadratic potential, as in eqn. \eqref{eq:GeneralPot} above, so we can find the frequency.
Define $\tilde E = E - U(x_0)$, which is also  conserved because $U(x_0)$ is constant.
Then the following is our generalized energy equation,
$$\tilde E=\frac12 m\dot \varepsilon^2+\frac12 U''(x_0)\varepsilon^2.$$

But this has exactly the same $U(\varepsilon) \propto \varepsilon^2$ (i.e., quadratic form) that we saw for the spring or the pendulum.
That is, it is the energy of a SHO! 
By our method for determining frequencies for SHO's, we have
$$\omega^2=U''(x_0)/m.$$

This form for the oscillation frequency will be true for motion in any force near a local minimum in the potential.
**You can (almost) always find a range of $x$ values about the minimum such that the potential looks like a quadratic about the minimum**, through a Taylor series expansion.
The "almost" is for the case where $V''(x)=0$.
Then this term disappears and you get a higher order dependence.  

**Worked example: Morin 5.9 p.174**: A particle moves under the influence of a potential $U(x)=-Cx^n\eexp{-ax}.$
Find the frequency of small oscillations around the equilibrium point.

**Solution:** First, we need to find the equilibrium point, which happens when $F = -\odel{U}x = 0$.
$$U'(x) = -C\eexp{-ax}x^{n-1}(n-ax)=0\quad\text{when}\quad x=x_e=n/a.$$
To find the frequency, we then compute the second derivative of $U$ at $x_e$, i.e.,
$$U''(x_e)=C\eexp{-n}n^{n-1}/a^{n-2}$$
and so 
$$\omega=\sqrt{\frac{C\eexp{-n}n^{n-1}}{ma^{n-2}}}.$$
The frequency is well defined as long as $U''>0$, i.e. as long as $U(x_e)$ is a minimum.

# The view from phase space

Conservation of energy $E = mv^2/2 + U(x)$ is written in terms of $v$ and $x$.
We can imagine plotting isosurfaces of $E$ as functions of $v$ and $x$ in the format of a contour plot.
In this plot, the trajectories of particles will go along surfaces of constant energy.
Let's consider three examples.

## SHO

A spring with $U = kx^2/2$ has energy
$$E=\frac12 mv^2+\frac12 kx^2$$
and describes an ellipse for each value of $E>0$.
The points $x=\pm\sqrt{2E/k}$ are the turning points where $v=0$.

We might have expected this behaviour, since we already determined that sinusoidal motion corresponds to ellipses in phase space.

To plot this, we use the function `numpy.meshgrid` to create a grid of values of $x$ and $v$.

```python
def E_SHO_phaseplot():
    ftsz = 12  # font size
    x = np.linspace(-5, 5, 100)  # position
    v = np.linspace(-5, 5, 100)  # velocity

    [xg, vg] = np.meshgrid(x, v)  # gridded values of x and v

    m = 1.0  # mass
    k = 2.0  # stiffness constant

    U = 0.5*k*x**2  # for the linear plot of U vs x
    Eg = 0.5*m*vg**2 + 0.5*k*xg**2  # E for the gridded contour plot

    plt.figure(figsize=(5, 6))
    plt.subplot(3, 1, 1)
    plt.plot(x, U)
    plt.xlim([x[0], x[-1]])
    plt.grid('on')
    plt.ylabel('$U(x)$ [J]', fontsize=ftsz)

    plt.subplot(3, 1, (2, 3))
    plt.contourf(xg, vg, Eg)  # make contour plot on grid xg, vg with values energy
    plt.colorbar(orientation='horizontal')  # add colourbar
    plt.contour(xg, vg, Eg, colors='k')  # outline the contours in black
    plt.xlim([x[0], x[-1]])
    plt.ylim([v[0], v[-1]])
    plt.grid('on')
    plt.title('Energy in phase space [J]', fontsize=ftsz)
    plt.xlabel('$x$ [m]')
    plt.ylabel('$v$ [m/s]')

    plt.suptitle('Energy of a simple harmonic oscillator', fontsize=ftsz+2)
    plt.tight_layout()
    return
```

```python
E_SHO_phaseplot()
```

(*Note: the $x$-axes match.*) The top panel shows the potential energy $U(x)$, and the bottom panel shows the total energy $E(x, v)$. The oscillation occurs around the energy minimum, proceeding along energy contours in a clockwise direction.

## Projectile motion

For projectile motion we have
$$E=\frac12 m v^2 + mgz,$$
where $v=\dot z$.
This is the equation for a family of parabolas lying on their side.

```python
def E_proj_phaseplot():
    ftsz = 12  # font size
    z = np.linspace(-50, 50, 100)  # [m] position
    v = np.linspace(-50, 50, 100)  # [m/s] velocity
    [zg, vg] = np.meshgrid(z, v)  # gridded values of z and v
    m = 1.0  # [kg] mass
    g = 9.8  # [m/s2] gravitational acceleration

    U = m*g*z  # for U vs x plot

    Eg = 0.5*m*vg**2 + m*g*zg  # for contour plot

    plt.figure(figsize=(5, 6))
    plt.subplot(3, 1, 1)
    plt.plot(z, U)
    plt.xlim([z[0], z[-1]])
    plt.ylabel('$U(z)$ [J]', fontsize=ftsz)

    plt.subplot(3, 1, (2, 3))
    plt.contourf(zg, vg, Eg) #plot contours of energy on grid in zg,vg
    plt.colorbar(orientation='horizontal')  # add colorbar
    plt.contour(zg, vg, Eg, colors='k')  # plot black contour lines
    plt.xlim([z[0], z[-1]])
    plt.ylim([v[0], v[-1]])
    plt.grid('on')
    plt.title('Energy in phase space [J]')
    plt.xlabel('$z$ [m]', fontsize=ftsz)
    plt.ylabel('$v$ [m/s]', fontsize=ftsz)

    plt.suptitle('Energy of projectile motion', fontsize=ftsz+2)
    plt.tight_layout()
    return
```

```python
E_proj_phaseplot()
```

(*Note: the $x$-axes match*) Here, there is no energy minimum or oscillatory motion, so the trajectories just correspond to the rising and falling motion of vertically projected particles.
There is a single turning point for each trajectory, corresponding to the top of the trajectory where $z=E/mg$.

## Simple non-linear pendulum

The energy for a general pendulum is
$$E=\frac12 m\ell^2\dot\theta^2 + mg\ell(1-\cos\theta).$$

In this case, the potential energy is periodic in $\theta$. Here, we plot the energy in ($\theta, \dot\theta$) phase space.

```python
def E_pendulum_phaseplot():
    
    ftsz = 12  # font size
    m = 1.0  # [kg] mass
    g = 9.8  # [m/s] gravity
    l = 1.0  # [m] pendulum length

    #theta and dottheta values
    th = np.linspace(-2*np.pi, 2*np.pi, 500)
    dotth = (g/l)**0.5*np.linspace(-2.5, 2.5, 500)

    [thg, dotthg] = np.meshgrid(th, dotth)  # gridded values

    U = m*g*l*(1-np.cos(th))  # for PE vs theta plot

    Eg = 0.5*m*l**2*dotthg**2 + m*g*l*(1-np.cos(thg))  # for contouring

    plt.figure(figsize=(5, 6))
    plt.subplot(3, 1, 1)
    plt.plot(th, U/m/g/l)
    plt.xlim([th[0], th[-1]])
    plt.grid('on')
    plt.ylabel(r'$U(\theta)/(mg\ell)$', fontsize=ftsz)

    plt.subplot(3, 1, (2, 3))
    plt.contourf(thg, dotthg, Eg/(m*g*l), np.arange(0,6,0.5))  # plot contours
    plt.colorbar(orientation='horizontal')  # add colorbar
    plt.contour(thg, dotthg, Eg/(m*g*l), np.arange(0, 6, 0.5), colors='k')  # black contour overlay
    plt.contour(thg, dotthg, Eg, [2*m*g*l], colors='r', linestyles='--')  # separatrix in dashed red
    plt.xlim([th[0], th[-1]])
    plt.ylim([dotth[0],dotth[-1]])
    plt.grid('on')
    plt.title(r'$E/(mg\ell)$ in phase space')
    plt.xlabel(r'$\theta$ [rad]')
    plt.ylabel(r'$\dot\theta$ [rad/s]')

    plt.suptitle('Energy of pendulum', fontsize=14)
    plt.tight_layout()
    return
```

```python
E_pendulum_phaseplot()
```

(*Note: the $x$-axes match.*) We can identify two regimes:

* For $E<2mg\ell$, the energy of the oscillator is not sufficient for the pendulum to roll, i.e., to get "over the top".
    The pendulum motion is oscillatory and bounded by the turning points $\theta$ such that $\cos\theta=1-E/(mg\ell)$ and therefore $\dot\theta=0$.
    And for very small oscillations, the bottom of the wells are approximately elliptical.
* For $E>2mg\ell$, we have rolling motion.
    A pendulum started from $\theta=0$ with $E>2mg\ell$ will "escape" its local potential well and undergo rolling motion.
    When $\theta=0$, $E= mv^2/2$.
    This implies that $v_0$ needs to be greater than the escape velocity $2\sqrt{g\ell}$ for rolling motion to occur.
    *Practice: what would the escape velocity be for a pendulum started at $\theta=\pi/4$?*

The boundary between the two types of motion is the curve $E=2mg\ell$ or $E/(mg\ell)=2$: this is known as the *separatrix*, highlighted in dashed red in the picture above.
Notice how the "$\times$" crossings of the separatrix occurs at the local maxima of the potential energy: these are the unstable positions $\theta=\pm\pi$ of the pendulum.
In HW2, you are to demonstrate that the phase space trajectories in this case are hyperbolas.
These crossings represent the two asymptotes of the hyperbola.
