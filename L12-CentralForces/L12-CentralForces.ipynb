# Introduction

A central force (CF) points radially from origin and depends only on the distance from it $r=|\vec r|$.
That is, a force $\vec F$ is is said to be central if 
$$\vec F = F(r)\unit{r}.$$
Equivalently, we can define a central force as one whose potential only depends on $r$, but we need to discuss why such a force would be conservative first.

CF equations are relevant to a very large class of problems, for example planetary motion about a large star, scattering problems involving interactions between two particles.
Morin derives the central force equation of motion from Lagrangian formalism, but this isn't necessary.
We will first consider conservation of energy in 3D, and then focus on the central force case.

# Central forces conserve energy

## Potential energy of a central force

Recall from lecture 6 what we mean by "a force conserves energy": we mean that we can define a potential energy associated with that force that can store or release energy.
It does not mean that energy is conserved, because it does not preclude the existence of other forces such as friction that aren't conservative.

In lecture 6, we introduced the two necessary and sufficient conditions for a force to be conservative:

1. That the force only depends on position $\vec r$, not time or velocity: $\vec F = \vec F(\vec r)$.
2. That the force can be expressed as the gradient of a potential: $\vec F = -\grad{U}$.

In the last expression, the minus sign is necessary to say things like "$E = K+U$" or "Equilibrium points are at the bottom of potential wells" whenever appropriate. (We'd need to flip all signs otherwise.)

We recall that in Cartesian coordinates,
$$\grad{f} = \pde{f}x \unit x + \pde{f}y \unit y + \pde{f}z \unit z,$$
and in less dimensions, you can ignore appropriate terms.
    For example, if the motion happens in the $(x, z)$ plane, you can write $\grad{f} = (\partial f/\partial x)\unit x + (\partial f/\partial z) \unit z$.
In polar coordinates, the gradient of a function $f(r)$ is
$$\grad{f} = \pde{}{r}\unit r + \frac1r\pde{}\theta \unit \theta.$$
So, if the force is central, there is no $\theta$-dependence and and $\grad{f} = (\odel{f}r)\unit r$.
Defining $U(r)$ such that
$$F(r) = -\ode{U}r,$$
we have our potential energy. We can then say that **all central forces are conservative**.

Even though we will study motion happening on a 2D plane (think orbits), we are in the 1D case that we focused on in lecture 6 because of the "central" character of $\vec F$.

The potential energy also only depends on $r$, and we calculate it by integrating $F$ between a reference radius $r_0$ and our radius $r$:
\begin{align*}
U(r) - U_0 & = -\int_{r_0}^{r} F(r') \di r',
\end{align*}
with $U_0 = U(r=r_0)$.
$U_0$ is as arbitrary as $r_0$ is, but in practice, we try to find a $r_0$ such that $U_0 = 0$.

**Example:** Gravitational force between two masses $M$ and $m$:
$$\vec F = -\frac{GMm}{r^2} \unit r \thrfor U(r) - U_0 = GMm\int_{r_0}^r \frac{\di r'}{{r'}^2} = GMm\left(\frac{1}{r_0} - \frac1{r}\right).$$
If we choose $r_0 \to \infty$ and $U_0 = 0$, we have the much more convenient expression $U(r) = -GMm/r$.

**Example:** any 1D force that depends only on position, like the spring force, is central. We choose $r_0$ as the equilibrium point.

## Rate of change of the mechanical energy

So, the force might be pointing radially, but the motion could still happen in any of the three dimensions (think orbits).
For motion in 3D, kinetic energy (KE) is
$$K = \frac12 mv^2= \frac12 m\vec v\cdot \vec v,$$
where $\vec v = (\dot x, \dot y, \dot z)$.
The rate of change of KE is
$$\dot K = \ode{}t \left(\frac12 m\vec v \cdot \vec v \right) = \frac m2\left(\dot{\vec v} \cdot \vec v + \vec v \cdot \dot{\vec v}\right) = \ode{(m\vec v)}t \cdot \vec v = \vec F\cdot\vec v,$$
by Newton's second law, and this is true for any force, central or not.

Also for a general force, the rate of change of potential energy following a particle is
$$\dot U = \ode{U(\vec r)}t = \pde{U}x \dot x + \pde{U}y \dot y + \pde{U}z \dot z = \vec v \cdot \grad U.$$

In the case of a conservative force where $\grad U = -\vec F$, we have
$$\dot U = -\vec F \cdot \vec v = -\dot K \thrfor \boxed{\dot E = 0},$$
where
$$E=\frac12 mv^2+U(\vec r).$$
This is true for any conservative force, and therefore for the special case of central forces.

# Central forces conserve angular momentum

Let us introduce the angular momentum of a particle of mass $m$ in a central field, 
$$\vec L = \vec r \times \vec p,$$
where $\vec p = m\vec v$ is the momentum of the particle.

$\vec L$ has some useful properties, some of which we now list and demonstrate.

### If a particle is subject to a central force only, $\vec L$ is conserved

Indeed,
$$\dot L=\dot{\vec r}\times\vec p + \vec r\times\dot{\vec p}=\vec v\times (m\vec v) + \vec r \times \vec F=0,$$
where we used $\dot {\vec p} = \vec F$ and the fact htat $\vec F \parallel \vec r$ if the force is central.
Indeed, recall that if two vectors $\vec v_1$ and $\vec v_2$ are parallel, which is the case for both terms in the equation above, then $\vec v_1 \times \vec v_2 = 0$.

The vectors $\vec r$ and $\vec p$ are both in the plane of motion, by contruction/definition.
By construction, $\vec L \perp \vec r, \vec p$ and so $\vec L$ is perpendicular to the plane of motion.

### If $\vec L$ is conserved, the plane of motion is fixed

This makes sense intuitively: once a particle starts in a particular plane of motion, there is no component of force to kick it out of that plane.
In other words, if the particle were to deviate from its initial plane of motion, in which direction would it go? Would it go above the plane, or would it go below it?
We can't even define what is above and below because there is nothing to break that symmetry.

If the plane of motion is fixed, we can define polar coordinates in the plane of motion. See figure above for a sketch, and let the third dimension be carried by $\unit z = \unit r \times \unit \theta$.
We can then define $L$ such that $\vec L = L\unit z$.

### Dynamics under conservation of $\vec L$

Let's remind ourselves of some expressions from the lecture on polar coordinates.
The velocity is
$$\vec v \equiv v_r \unit r + v_\theta\unit\theta = \dot r \unit r + r\dot\theta\unit\theta,$$
and $\vec F = m\vec a$ now reads
$$F_r = F = m\left(\ddot r -r\dot\theta^2\right)=-\ode Ur \andeq F_\theta = 0 = m\left(2\dot r \dot \theta + r\ddot\theta\right).$$

We can now derive equations (7.6) of Morin without Lagrangians:
\begin{equation}
m\left(\ddot r - r\dot\theta^2\right)=-\ode Ur
\label{eq:reqn}
\end{equation}
for the $\unit r$ direction and
\begin{equation}
m\left(2\dot r\dot\theta+r\ddot\theta\right) = \frac1r \ode{(mr^2\dot\theta)}{t}=0
\label{eq:thetaeqn}
\end{equation}
for the $\unit \theta$ direction.

Eq. \eqref{eq:thetaeqn} is statement of angular momentum conservation:
$$\vec L=\vec r\times \vec p=m\vec r \times\vec v = m\left(r\unit r\right) \times\left(\dot r \unit r + r\dot\theta\unit\theta\right) = mr^2\dot\theta\left(\unit r \times\unit \theta\right)\equiv L \unit z,$$
where 
\begin{equation}
L = mr^2\dot\theta,
\label{eq:Lexpr}
\end{equation}
which can be positive or negative depending on $\dot\theta$.

Thus, eq. \eqref{eq:thetaeqn} says that the angular momentum $L=mr^2\dot\theta$ is conserved following the motion.

*Just like we call both $\vec F$ and $F$ "the force" when there is no ambiguity about the orientation, we might refer to both $\vec L$ and $L$ as "the angular momentum", because for central forces, the plane of motion is fixed and therefore, there is no ambiguity that $\unit z$ defines its orientation.*

# Solving the central field problem

We will use energy and momentum conservation to solve the CF problem, in the way we solved general motion in a one-dimensional potential.

## The effective potential

We can rewrite Newton's second law along $\unit r$ (eq. \ref{eq:reqn}) using conservation of $L$.
First, $r\dot\theta^2=v_\theta^2/r$ can be expressed as the gradient of a function of $r$ thanks to eq. \eqref{eq:Lexpr}:
$$\dot\theta = \frac{L^2}{mr^2} \thrfor mr\dot\theta^2=\frac{L^2}{m r^3}=-\ode{}{r}\left(\frac{L^2}{2m r^2}\right).$$
We've used $L$ conservation to turn this component of the acceleration into a function of $r$.

We can move $-mr\dot\theta^2$ to the right hand side of eq. \eqref{eq:reqn}, which will become an "effective" force term.
\begin{align}
m\ddot r  & = -\ode Ur -\ode{}{r}\left(\frac{L^2}{2mr^2}\right), \oreq \nonumber \\
m \ddot r & = -\ode {U_{\text{eff}}}r, \label{eq:mar}
\end{align}
where
$$\boxed{U_{\text{eff}}(r)=U(r)+\frac{L^2}{2mr^2}}$$
is the "effective potential".

This equation is exactly the same as conservative motion in a 1D potential that we saw in the lecture on energy, that is, $m\ddot x = -\odel Ux$.
Eqs. \eqref{eq:reqn} and \eqref{eq:mar} (they are the same) conserve energy
\begin{equation}
E = \frac12 m \dot r^2 + U_{\text{eff}}(r)
\label{eq:E_Ueff}
\end{equation}
and its solution is characterised by
$$t-t_0=\pm \sqrt{\frac m2}\int_{r_0}^r\frac{\di r'}{\sqrt{E-U_{\text{eff}}(r')}}.$$
This gives us a way of automatically finding the distance from the origin $r(t)$.

So, not only are $F$ and $U$ one-dimensional, but we can solve the whole 2D motion by solving intermediate 1D problems.

To find the angle $\theta(t)$, we use conservation of $L$ again:
$$L=mr^2\dot\theta=const.\thrfor \theta-\theta_0 = \frac Lm\int^t_{t_0}\frac{\di t'}{r^2(t')}.$$

By the way, is the $E$ we wrote down in eq. \eqref{eq:E_Ueff} really the energy? Yes, but the kinetic and potential energies are not what you'd think if you looked at it too quickly. Indeed, the potential energy is still $U$, not $U_{\text{eff}}$, and
$$K = \frac12 mv^2=\frac12 m\left(v_r^2+v_\theta^2\right).$$


Now $v_r=\dot r$ and $v_\theta=r\dot\theta$ so
$$K = \frac12 m \left(\dot r^2+r^2\dot\theta^2\right) = \frac12 m\left(\dot r^2+\frac{L^2}{2mr^2}\right)$$
and therefore, $E=K+U$ can be written
$$E=\frac12 mv^2 + U(r)=\frac12 m\dot r^2+\frac{L^2}{2mr^2} + U(r)=\frac12 m\dot r^2 + U_{\text{eff}}(r).$$

## Effective force

We're going to find that it is very useful to think of $U_{\text{eff}}$ as an actual potential and $-\odel{U_{\text{eff}}}r$ as a force.
What do they look like?

Effective radial force is
$$F_{\text{eff}}=-\ode{U_{\text{eff}}}r = \underbrace{\frac{L^2}{mr^3}}_{\text{centrifugal}\\ \text{force}}-\ode{U}r.$$ 
The first term is always positive, and therefore, it corresponds to a repulsive force.
And because it is about a central motion, we describe it as "centrifugal", but it is also sometimes called the "angular momentum barrier", because it prevents objects to get too close to the origin.

As for the second term, it can be attractive or repulsive, depending on the force.

* **Repulsive force:** $F(r)=-\odel Ur>0$, so $-\odel{U_{\text{eff}}}r>0$.
    All trajectories are unbounded: Eventually, $r\to \infty$.
    For example, the Coulomb force applied to two charges of the same sign is repulsive.
* **Strongly attractive force:** $F(r)<0$ and strong enough to overcome centrifugal force term everywhere. 
* **Moderately attractive force:** This is the most interesting: we can get orbits around the origin. 
    In these cases, $U_{\text{eff}}(r)$ looks like it has a potential well, meaning that there is a well-defined orbit radius, but also that the radial position can oscillate around it if the deviations aren't too strong.
    Also, don't forget that $\theta$ changes, and that oscillations in $r$ correspond to variations in distance from origin in orbital plane.

```python
rad = np.linspace(0.00001, .1, 1024)
L = 1.  # angular momentum
m = 1.  # mass
CFF = 0.5*rad**-2  # centrifugal force
Urep = 100*rad**-1  # repulsive potential
Uatt_m = -100*rad**-1  # moderately attractive potential
Uatt_s = -0.1*rad**-2.5  # strongly attractive potential
```

```python
def plot_potentials():
    
    plt.figure(figsize=(8, 6))
    
    plt.subplot(3, 1, 1)
    plt.plot(rad, CFF, ':', label='$1/(2r^2)$')
    plt.plot(rad, Urep, '--', label='$U(r) = 100/r$')
    plt.plot(rad, Urep+CFF, label=r'$U_{eff}(r)$')
    # plt.xlabel('$r$')
    plt.ylabel('repulsive')
    plt.xlim([0., max(rad)])
    plt.ylim([0., 10000])
    plt.grid()
    plt.legend(loc='upper right')
    
    plt.subplot(3, 1, 2)
    plt.plot(rad, CFF, ':', label='$L^2/(2mr^2)$')
    plt.plot(rad, Uatt_s, '--', label='$U(r) = -0.1/r^{2.5}$')
    plt.plot(rad, Uatt_s+CFF, label=r'$U_{eff}(r)$')
    # plt.xlabel('$r$')
    plt.ylabel('strongly attractive')
    plt.xlim([0., max(rad)])
    plt.axhline(0., c='k')
    plt.ylim([-10000, 10000])
    plt.grid()
    plt.legend(loc='upper right')

    plt.subplot(3, 1, 3)
    plt.plot(rad, CFF, ':', label='$L^2/(2mr^2)$')
    plt.plot(rad, Uatt_m, '--', label='$U(r) = -100/r$')
    plt.plot(rad, Uatt_m+CFF, label=r'$U_{eff}(r)$')
    plt.xlabel('$r$')
    plt.ylabel('moderately attractive')
    plt.xlim([0., max(rad)])
    plt.axhline(0., c='k')
    plt.ylim([-10000, 10000])
    plt.grid()
    plt.legend(loc='upper right')
    
    plt.tight_layout()
```

### Circular orbits

These are special cases of the latter.
Suppose we can find the radius $r_0$ where the centrifugal and central forces are exactly balanced.
If we start with $\dot r = 0$, the particle will stay at this radius for all time.
So the orbits are *circles*, and $\dot\theta = L/(mr_0^2) = \omega_{\text{orbit}}$ is the orbital frequency and is constant.
The period of the orbit is
$$T=\frac{2\pi mr_0^2}{L}.$$

In summary, the central force problem is completely integrable using conservation of angular momentum and energy, respectively 
$$L=mr^2\dot\theta \andeq E=\frac12 m\dot r^2+U_{\text{eff}}(r),$$
where the effective potential is
$$U_{\text{eff}} = U(r)+\frac{L^2}{2mr^2}.$$

**Worked example:** A particle of mass $m$ moves under an attractive potential $$U(r)=\frac 15 Cr^5,$$
where $C$ is a constant.
For what energy will the motion be circular, what is the radius of the circle, and what is the orbital period?

For this case,
$$U_{\text{eff}}(r) = \frac 15 Cr^5+\frac{L^2}{2mr^2}.$$
Now the force
$$F_{\text{eff}} = -\ode{U_{\text{eff}}}r = \frac{L^2}{mr^3} - Cr^4.$$
$F_{\text{eff}}=0$ at $r_0=\left[L^2/(Cm)\right]^{1/7}.$
The orbital frequency is
$$\omega_{\text{orbit}}=\frac{L}{mr_0^2} = \frac{L}{m\left[L^4/\left(C^2 m^2\right)\right]^{1/7}} = \left(\frac{L^3C^2}{m^5}\right)^{1/7},$$
and the orbital period is $2\pi/\omega_{\text{orbit}}.$


```python
def Crpower5():
    # Plot for C = L = m = 1
    rad = np.linspace(0.00001, 2., 1024)
    U = 0.2*rad**5
    CFF = 0.5*rad**-2
    Ueff = CFF + U
    plt.plot(rad, Ueff)
    plt.xlim([0., max(rad)])
    plt.ylim([0, 7])
    plt.axvline(1., c='k', ls='--')
    plt.xlabel('$r/r_0$')
    plt.ylabel("$U_{eff}$")
    plt.grid()
```

```python
Crpower5()
```

Now what happens if we perturb this orbit? If $U''_{\text{eff}}(r_0)>0$, i.e. $U_{\text{eff}}$ is a local minimum, we can get oscillations in $r$.
These will show up as variations in the distance from the origin as the particle orbits.

For small amplitude perturbations, these oscillations will exhibit SHM in $r$.
To see this, remember the energy method:
$$E=\frac12 m\dot r^2 + U_{\text{eff}}(r),$$
but this is not in the "small-deviations", "double-quadratic" form we saw in the lecture on energy yet.
Consider small deviations $q$ around $r_0$: $q = r-r_0 \ll r_0$.
A Taylor expansion of the effective potential around $q=0$ then yields
$$U_{\text{eff}}(r) = U_{\text{eff}}|_{r_0} + q\, \underbrace{U'_{\text{eff}}|_{r=r_0}}_{=0} + \frac12 q^2 U''_{\text{eff}}|_{r=r_0} + O(q^3).$$
Therefore, we have conservation of $$\tilde E=\frac12 m\dot q^2+\frac12 \left.U''_{\text{eff}}\right|_{r=r_0}q^2.$$ 

Near the minimum of $U_{\text{eff}}$, we have small oscillations  $q(t)=A\cos(\omega t+\phi)$ with $\omega=\sqrt{\left.U''_{\text{eff}}\right|_{r=r_0}/m}$.

**Worked example:**
For $U=\frac15 Cr^5$, find the frequency of radial oscillations.
Are the resulting orbits open or closed?

$$\left.U''_{\text{eff}}\right|_{r=r_0} =7 C r_0^3=7 m\omega_{\text{orbit}}^2,$$
where we used 
$$Cr_0^7 = \frac{L^2}m = \frac{m^2 r_0^4 \omega^2_{\text{orbit}}}{m}.$$ 
So
$$\omega = \sqrt 7\omega_{\text{orbit}} \approx 2.6\omega_{\text{orbit}}.$$

Is this orbit closed? No, the particle crosses the circle several times but doesn't return to its original point. Generally, the orbits will only be closed if the orbital period is commensurate with the radial oscillation period.


```python
def Crpower5_oscillations():
    worbit = 1.  # orbital angular frequency
    wosc = 7**.5 * worbit
    time = np.linspace(0, 10*2*np.pi/worbit, 4096)  # ten orbits
    theta = worbit*time
    rad = 1. + 0.1*np.cos(wosc*time)  # exaggerate the amplitude a little

    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.scatter(theta, rad, c=theta, s=1.)  # colour-coded with time: blue=early, yellow=late
    ax.set_rmax(1.15)
    # ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
    ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
    ax.grid(True)

    plt.show()
```

# Kepler's laws

Let's shift our attention to what was arguably the main motivation to the development of mechanics during the [early modern period](https://en.wikipedia.org/wiki/Early_modern_period): the study of orbits around massive objects.
Kepler's work predated Newton's by a few decades, and was one of his main inspirations.
Kepler's laws are the following (shamelessy lifted from [Wikipedia](https://en.wikipedia.org/wiki/Newton%27s_law_of_universal_gravitation)):

> 1. The orbit of a planet is an ellipse with the Sun at one of the two foci.
  2. A line segment joining a planet and the Sun sweeps out equal areas during equal intervals of time.
  3. The square of a planet's orbital period is proportional to the cube of the length of the semi-major axis of its orbit.

In order of generality:

* The second law is the first one I will derive, because it is general to central forces.
* The first law only applies to inverse-square laws, and in certain conditions that all planets happen to meet.
    That is, they apply to Newton's law of universal gravitation as originally intended, but could also apply to Coulomb's law for oppositely-charged particles.
* And finally, the third law applies to all objects in the orbit of one, large body like the Sun.

In what follows, I assume that the origin of the central force is staying put at $r=0$.
In reality, it is not the case, and all objects are supposed to do a dance around the centre of mass of the system.
To address it, we would need to do one of two things: 

1. define the centre of the coordinate system to be on one of the masses, but accept that this centre would describe a circle if viewed from a Galilean reference frame, and we would need to add non-inertial corrections to the accelerations; or
2. define our origin to be the centre of gravity of the system, but we would need to rewrite the forces a little.

Here, like Kepler and Newton, we are imagining a Sun-Earth-type system, where the mass of one object is so much larger than that of the other, it barely moves. Bottom line is, we don't have to worry about the above.

## Kepler's second law

This one applies to *any* central force because it directly derives from conservation of angular momentum.

Consider a line segment joining the particle and the origin of the central force.
During an infinitesimal time $\di t$, the segment's azimuthal coordinate increments by $\di 
\theta$. (Cf. sketch below.)

```python
%%html
<img src="08-central-kepler2.png" width="300">
```

The area swept out that subtends angle $\di\theta$ during $\di t$ is $\di \mathcal A = r (r \di\theta)/2$ for vanishing $\di t$.
Thus,
$$\ode{\mathcal A}t = \frac12 r^2 \dot\theta = \frac{L}{2m}.$$
Therefore, if $L$ is conserved, so is $\mathcal A$.

In what follows, I will introduce the specific angular momentum $h = L/m$, so that
\begin{equation}
r^2\dot \theta = h.
\end{equation}

## Kepler's first law

### Foreword

Back to the first law.
In a central field, equal areas are swept out in equal times.
This means that the particle will move faster as it approaches the origin.
We wouldn't need such a law if the orbits were circular, because we'd just have $\dot \theta = constant$.
So, there have got to be more complicated orbits than just circles.

I find the derivation of Kepler's first law less interesting than many we've seen during the rest of this course.
My reason is that most of the derivation techniques we've seen (separation of variables, adopt-an-exponential, linear algebra, Fourier series...) are universal. 
As in, you saw them in this course, and you will see them again, over and over.
But the applicability of this particular derivation is confined to Keplerian dynamics, and won't make you better physicists. (Others might disagree, but I think so.)
It is a good exercise about how to manipulate equations, though, so, treat it as a worked example.

The general strategy is the following: because we are looking for the shape of the trajectories, we are only interested in spatial information.
In other words, we want to know $r(\theta)$ instead of $r(t)$ and $\theta(t)$, i.e., we will eliminate time from all of our equations.

### The Binet equations

The Binet equations state that for central forces,
\begin{equation}
    v^2 = h^2\left[\left(\ode{y}{\theta}\right)^2 + y^2\right] \andeq a_r = -h^2 y^2 \left(\oden2{y}{\theta} + y\right),
    \label{eq:Binet}
\end{equation}
where $v^2$ is the square of the velocity magnitude and $y = 1/r$.
Those two equations look a little random, and me, dropping them without justification certainly isn't helping, but they are useful enough that they have their own name.

***

*In the interest of completeness, here is the derivation. You may skip it.*

*Recall that in polar coordinates,*
$$\vec v = \dot r \unit r + r\dot\theta \unit\theta \thrfor v^2 = \dot r^2 + r^2 \dot\theta^2.$$
$$a_r = \ddot r - r \dot\theta^2,$$
*and recall that the game is to make all mentions of time disappear. It is possible, simply because all variables are univariate functions of time, and therefore $\theta$ is an unambiguous measure of time.*
*In practice, we do*
\begin{align*}
\dot\theta & = h/r^2 = hy^2,\\
\dot r & = \ode{r}t = \ode{r}{\theta}\ode{\theta}{t} = \frac{h}{r^2}\ode{r}{\theta} = h y^2 \left(-\frac{1}{y^2}\ode{y}{\theta}\right) = -h\ode{y}{\theta},\\
\ddot r & = -h\ode{}{t}\left(\ode{y}{\theta}\right) = -h\oden2{y}{\theta}\dot\theta = -h^2 y^2\oden2{y}{\theta}.
\end{align*}

*Collect all of these terms and equations \eqref{eq:Binet} follow.*

***

The derivation for the Binet equations is purely kinematic, and does not make any assumption other than the force being central, when we used conservation of angular momentum in the form $\dot \theta = h y^2$.

### Inverse square forces

Let us now introduce the inverse square force assumption, i.e.,
$$F = -\frac{\alpha}{r^2} = -\alpha y^2 \oreq U = -\frac{\alpha}r = -\alpha y.$$
For example, for Newton's law of gravity between two masses $M$ and $m$, $\alpha=GMm$.
Another example is Coulomb's law of electrostatics between two electric charges $Q$ and $q$, $\alpha = -Qq/(4\pi\varepsilon_0)$, where $\varepsilon_0$ is the permittivity of vacuum.

The second Binet equation is now
$$a_r = \frac{F}m = -\frac{\alpha y^2}m = -h^2 y^2 \left(\oden2{y}{\theta} + y\right),$$
or, assuming $y\neq 0$ ($r$ finite),
$$\oden2{y}{\theta} + y = \frac{\alpha}{mh^2} = \frac{m\alpha}{L^2}.$$

This is a classic ODE, and we've actually seen worse: it's like the SHO with $\omega_0^2 = 1$.
Its solution is the sum of a particular solution, $y_p = \alpha/(mh^2)$ being the easiest one, and of the general solution of the homogeneous ODE, $y_h = A\cos(\theta + \phi)$, where $A$ and $\phi$ are determined via initial conditions.
That is,
$$y = \frac{m\alpha}{L^2} + A \cos(\theta + \phi) = \frac{1 + e\cos(\theta + \phi)}{\ell},$$
where $\ell = L^2/(m\alpha)$ and $e = \ell A$.

Let us rotate our polar coordinates in order to write $\phi=0$, and our final solution is
\begin{equation}
\boxed{r = \frac{\ell}{1 + e\cos\theta}}.
\end{equation}

When $\theta=0$, $r=r_0 = \ell/(1+e)$.
At this point, we also have $\odel{r}{\theta} = 0$ because the function is even around $\theta = 0$.
That is, $\dot r = 0$ at $\theta=0$ and the energy is
$$E = \frac{L^2}{2 m r_0^2} - \frac{\alpha}{r_0}.$$
We could invert this expression to obtain
\begin{equation}
e = \sqrt{1 + \frac{2EL^2}{m\alpha^2}}.
\label{eq:ecc_E}
\end{equation}

This might be a good time to recall that we decided early on that $U\to0$ when $r\to 0$.
So, $E>0$ means that infinitely far away, the particle will still have some kinetic energy (unbounded motion), if $E<0$, the particle's motion is bounded to a potential well, and $E=0$ is the separatrix.
Let's see how these facts manifest themselves here.

### Three types of orbits

You may have recognized the equation for a conic section, with $e$ the eccentricity and $\ell$ the "semi-latus rectum", which is best defined graphically, as we will do below.
And if you are more familiar with the Cartesian expressions of conic sections, Morin does the conversion in his section 7.4.3.

Conic sections are split into three categories.

![Public Domain, https://commons.wikimedia.org/w/index.php?curid=587213](Eccentricity.png)

#### For $e>1$, we have hyperbolas.

These correspond to $E>0$ from equation \eqref{eq:ecc_E}.

They have their minimum distance from the origin when $\theta = 0$, i.e.,
$$r \geq \frac{\ell}{1+e},$$
but have no maximum distance.
That is, when $\cos\theta \to 1/e$, which corresponds to two angles, $r\to \infty$.
In other words, trajectories are unbounded.

Take one of the two foci on the figure below, say, the left one. If the force is attractive, the trajectory is described by the branch on the left.
But if the force is repulsive, the right branch is the one to select.

Note that for a repulsive force, $\alpha <0$ and so
$$E = \frac{L^2}{2mr_0^2} + \frac{|\alpha|}{r_0} >0.$$
That is, a charge of the same sign as the big particle will always move along a hyperbola.

![By Klaas van Aarsen - Own work, CC BY-SA 3.0, https://commons.wikimedia.org/w/index.php?curid=25261095](Conic_section_-_standard_forms_of_a_hyperbola.png)

#### For $e=1$, we have parabolas

These correspond to $E=0$: they are unbounded, but in parameter space, they would follow the separatrix.

So, $E=0$ corresponds to a category of orbit that starts infinitely far away with no kinetic energy, and then, for some reason, starts getting attracted (maybe it is far, but not infinitely, far away; or maybe it get a tiny kick towards the Sun).
It get speeds up towards the Sun, does a simple slingshot close to it, and goes back toward an infinitely far away, immobile state.

![By Klaas van Aarsen - Own work, CC BY-SA 3.0, https://commons.wikimedia.org/w/index.php?curid=25261094](Conic_section_-_standard_forms_of_a_parabola.png)

#### For e<1, we have ellipses

This is the $E<0$ case, and the one Kepler and Newton  were concerned with. In other words, **Kepler's first law applies to this state.**

The radius is bounded,
$$ \frac{\ell}{1 + e} \leq r \leq \frac{\ell}{1-e}.$$
The mimimum radius above is called the *perihelion* (in Greek: "peri": near; "helios": Sun), and the maximum radius is the aphelion ("apo-": away).

This case includes the circular orbit case, with $e=0$ or $E=-m\alpha^2/(2L^2)$.

![By Klaas van Aarsen - Created as a latex tikzpicture. Previously published: Not published before, CC BY-SA 3.0, https://commons.wikimedia.org/w/index.php?curid=25261046](Conic_section_-_standard_forms_of_an_ellipse.png)

This case is one of those described in the previous section, that of a circular orbit with perturbations, except that the perturbations have the exact same periodicity as the orbit, therefore, ellipses.

Indeed,
$$U_{\text{eff}} = \frac{L^2}{2 m r^3} - \frac{\alpha}{r} \thrfor U'_{\text{eff}} = -\frac{L^2}{mr^3} + \frac{\alpha}{r^2}.$$
The radius of the circular orbit is when the latter is zero, i.e., when
$$r_0 = \frac{L^2}{m\alpha} \thrfor \left.U''_{\text{eff}}\right|_{r=r_0} = \frac{3L^2}{mr_0^4} - \frac{2\alpha}{r_0^3}= \frac{\alpha^4 m^2}{L^6} \andeq \omega_{\text{orbit}} = \frac{L^2}{mr_0^2} = \frac{m \alpha^2}{L^3}.$$

The perturbations have angular frequency
$$\omega_{\text{pert}} = \sqrt{\frac{}{}}$$

## Kepler's third law

Since these laws apply in the elliptical case, let's stick to it for now.

Recall that the law states that "The square of a planet's orbital period is proportional to the cube of the length of the semi-major axis of its orbit."
In other words, is $a$ is the length of the semi-major axis of its orbit, and $T$ the orbital period, then $T^2/a^3 = constant.$

We have
$$a = \frac12\left(\frac{\ell}{1 + e} + \frac{\ell}{1 - e}\right) = \frac{\ell}{1 - e^2}.$$

The area of an ellipse is $\mathcal A = \pi ab$, where $b$ is the semi-major axis.
With a bit of trigonometry, we could show that
$$b = \frac{\ell}{\sqrt{1-e^2}} = a\sqrt{1-e^2} = \sqrt{a\ell},$$
and so the area of an ellipse is also $\pi\sqrt{a^3\ell}$.

We also know, from Kepler's second law, that over one orbital period, $\mathcal A = \pi \sqrt{a^3\ell} = LT/(2m)$, which, after squaring, yields,
$$\frac{T^2}{a^3} = \frac{4m^2\pi^2 \ell}{L^2}.$$
Also, recall that $\ell = L^2/(m\alpha)$ from when we first introduced it.
We then have
$$\frac{T^2}{a^3} = \frac{4m\pi^2}{\alpha}.$$

The law applies to any elliptical orbit.
But in the case of the solar system (or any solar system), where the situation is approximately that all planets orbit around the Sun of mass $M\gg m$, $\alpha = GMm$, and we have
$$\boxed{\frac{T^2}{a^3} = \frac{4\pi^2}{GM}}.$$
That is, this ratio does not depend on the planet, because its mass dropped out of it: all planets have this ratio in common.

Note that this would not work in a system of charged particles, even if quantum mechanics wasn't a thing, because $\alpha$ would not be proportional to $m$.

* Added to the central potential, this "centrifugal" potential contributes to an effective potential. This effective potential is sometimes only repulsive, sometimes only attractive, and sometimes it is repulsive close to the central point, and attractive away from it.
* Very standard principles from the lecture on energy, applying to a particle in our effective well, can help us derive basic results for the radius and period of the orbit, and perturbations about it.
* Kepler's second law, which states that an orbit sweeps equal areas in equal times (not quite... It's hard to state concisely and clearly), is actually very general to any central force problem.
* If the particle is in an inverse-square potential, the orbits are conic sections: unbounded hyperbolas at high energies, bounded ellipses at low energies, and parabolas as the separatrix.
* Ellipses fall into the "circular orbit + perturbation" category, but the inverse-square force makes the perturbations' frequency the same as the orbit frequency, hence the clean-looking orbits.
